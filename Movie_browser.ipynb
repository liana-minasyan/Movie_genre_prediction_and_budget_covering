```python
url="https://www.rottentomatoes.com/browse/dvd-streaming-all/"
```

```python
browser = webdriver.Chrome()
time.sleep(5)
browser.get(url)
time.sleep(5)
while True:
    some_element = browser.find_element_by_tag_name("html")
    some_element.send_keys(Keys.END)
    time.sleep(2)
    show_more = browser.find_element_by_css_selector('div[id="show-more-btn"]>button[class="btn btn-secondary-rt mb-load-btn"]')
    show_more.click()
    time.sleep(3)
```

```python
all_urls=[]
page=browser.page_source
response = TextResponse(url=url,body=page,encoding="utf-8")
one_page_urls = response.css('h3[class="lister-item-header]>a"').extract()
all_urls.extend(one_page_urls)
```
