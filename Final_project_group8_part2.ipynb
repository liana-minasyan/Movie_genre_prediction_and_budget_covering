```python
#regex functions for cleaning cpecific columns
def money(i):
    if len(re.findall("\d\smillion", str(i))) != 0:
        y = re.findall("\d\smillion", str(i))[0]
        return float(re.findall("\d+", str(y))[0])*1000000
    elif len(re.findall("\d+\.*\d\smillion", str(i))) != 0:
        y = re.findall("\d+\.*\d\smillion", str(i))[0]
        return float(re.findall("\d+.*\d", str(y))[0])*1000000
    elif len(re.findall('\d+\.*[\-|\–]\d+', str(i)))!= 0:
        x = (re.sub('[\-|\–]', '$', str(i)))
        try:
            return float(re.findall("(\d+\.*\d)\$", x)[0])
        except:
            pass
    elif len(re.findall("\d+.*\d", str(i))) != 0:
        u = re.sub('[,|\s]', '',(re.findall("\d+.*\d", str(i))[0]))
        return float(re.findall("\d+\.*\d",u)[0])
    elif len(re.findall("\d+.*\s\d", str(i))) != 0:
        return float(re.sub(' ', '',(re.findall("\d+\.*\d", str(i))[0])))
    else:
        return 0
```

```python
def clear_first(a):
    if len( re.findall("[A-Z][a-z]+\s+[A-Z]*[a-z]*", str(a))) > 0:
        return re.findall("[A-Z][a-z]+\s+[A-Z]*[a-z]*", str(a))[0]
    else:
        return 0
def clear_prod(a):
    if len(re.findall("[P|p]lainlist|[P|p]lain\slist", str(a)))!= 0:
        x  = re.sub("[P|p]lainlist|[P|p]lain\slist", "", str(a))
        return re.findall("[A-Z][\w*\'*a-z]+[\sA-Za-z]*", x)[0]
    if len( re.findall("[A-Z][\w*\'*a-z]+[\sA-Za-z]*", str(a))) > 0:
        return re.findall("[A-Z][\w*\'*a-z]+[\sA-Za-z]*", str(a))[0]
    else:
        return 0 
def clear_dist(a):
    if len(re.findall("[P|p]lainlist", str(a)))!= 0:
        x  = re.sub("P|plainlist", "", str(a))
        return re.findall("[A-Z][\w*\'*a-z]+[\sA-Za-z]*", x)[0]
    if len( re.findall("[A-Z][\w*\'*a-z]+[\sA-Za-z]*", str(a))) > 0:
        return re.findall("[A-Z][\w*\'*a-z]+[\sA-Za-z]*", str(a))[0]
    else:
        return 0
def clear(a):
    if len( re.findall("[A-Z][a-z]+\s+[A-Z][a-z]+", str(a))) > 0:
        return re.findall("[A-Z][a-z]+\s+[A-Z][a-z]+", str(a))
    else:
        return 0
def clear_runtime(a):
    #check if a is not NaN
    if (a==a and len(re.findall("\d+", a))!= 0):
        return float(re.findall("\d+", a)[0])
    else:
        return 0   
```
